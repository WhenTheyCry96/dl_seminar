```
def sum_many(*args):
    sum = 0
    for i in args:
        sum = sum + i
    return sum


print (sum_many(10, 20, 30, 50, 600))
```

```
def func_kwargs(*args, **kwargs):
    print (args)
    print (kwargs)


func_kwargs(1, 2, 3, name="parkSH", age=22)
```
