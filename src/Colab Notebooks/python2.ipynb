```
money = 1
print ("money = ", end=" ")
print(money, end=" ")
print(type(money))
if money == 1:
    print ("money == 1 (int)")
else:
    pass
if money is 1:
    print ("money is 1 (int)")
else:
    pass
```

```
money = 12345
print ("money = ", end=" ")
print(money, end=" ")
print(type(money))
if money is 12345:
    print ("money is 12345 (int)")
else:
    pass
```

```
money = "1"
print ("money = ", end=" ")
print(money, end=" ")
print(type(money))
if money == 1:
    print ("money = 1 (int)")
elif money == "1":
    print ("money = '1' (str)")
else:
    print ("none")
```

```
money = 1000000
print ("money = ", end=" ")
print(money, end=" ")
print(type(money))
# compare operator : > < >= <= == !=
if money > 10000:
    print ("money > 10000")
# and or not
if money > 10000 and money < 10e8:
    print ("10e4 < money < 10e8")
```

```
treeHit = 0
while treeHit < 10:
    treeHit = treeHit +1
    print("나무를 %d번 찍었습니다." % treeHit)
    if treeHit == 10:
        print("나무 넘어갑니다.\n")
```

```
num = 0
while num < 10:
    num = num +1
    if num % 2 == 0:
        continue
    print("num : %d" % (num))
```

```
coffee = 10
money = 300
# 무한 루프?
while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee -1
    print("남은 커피의 양은 %d개입니다." % coffee)
    if not coffee:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.\n")
        break
```

```
marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks:
    number = number +1
    if mark < 60: continue
    print("%d번 학생 %d 점으로 축하합니다. 합격입니다. " % (number, mark))
```

```
marks = [90, 25, 67, 45, 80]
number = 0
for number in range(len(marks)):
    if marks[number] < 60: continue
    print("%d번 학생 %d 점으로 축하합니다. 합격입니다." % (number+1, marks[number]))
```

```
marks = [90, 25, 67, 45, 80]
number = 0

for number, element in enumerate(marks):
    if element < 60: continue
    print("%d번 학생 %d 점으로 축하합니다. 합격입니다." % (number+1, element))
```
